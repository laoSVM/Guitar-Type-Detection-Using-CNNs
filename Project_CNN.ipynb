## Introduction/Motivation

When making the decision to learn a musical instrument, whether young or old, it could be challenging. There’s a lot to know and lots of people’s opinions. Looking for instruments at a store is one thing, there’s always or at least should be, a salesperson there to guide you through the choices. But when you venture on the web on your own it can be a little bit trickier. You are trusting that whatever site you are looking on has all their products labeled and categorized correctly. For the most part they are but there can be mistakes or lack of description. This would be much more the situation on a site where people can sell their own instruments, think a place like eBay. This is where a Convolutional Neural Network that is able to decipher what the instrument is based on the image would be very beneficial. E-commerce websites with thousands of listings would greatly benefit from this. Ones searching things like concert photos would be able benefit as well. To that end, we are going to look at a Convolutional Neural Network that can distinguish an electric guitar from an acoustic guitar.

In order to understand what we are looking at, a little background about the two guitar types is needed. Both types have a body, a neck with fretboard, a headstock with tuning pegs, bridge, and six strings. An Electric guitar has a solid wood body while an Acoustic guitar has a hollow body and a hole in the center. The strings lay over the fretboard and are fastened to the center of the guitar, at the bridge (below the pickups or hole), and headstock, where the strings are tuned with the tuning pegs. 

Both types serve the same purpose as an instrument, it's just the tone of the instrument is different. An Electric guitar is traditionally used in pop and rock music. It is required to be plugged into an amplifier to produce sound. The Electric guitar has 2-3 rectangular pickups on the body under the strings, these are essentially microphones that pick up the vibrations on the strings. It also has volume and tone knobs and a possible pickup switch lever. An Acoustic guitar’s sound is produced by the whole and hollowness of the body. They amplify the vibration of the strings and produce the sound. An Acoustic guitar is used a lot on folk and country music. 



## Literature Review

We were able to find three projects online that did very similar CNNs. While all a little different there was a lot we were able to learn from them. While we were hoping to find more we understand this project is a little specific. In looking for more research most of the search results were about CNNs used to detect what instrument was being played in an audio recording, what was the dominant instrument being played in an audio recording, detecting the notes played in an audio recording and transcribing them, and detecting emotion in an music recording. While these are related to our project they were nonetheless extremely fascinating.

The first project we looked at was “Electric guitar classification using CNNs”  by Debadri Sengupta. Unlike our project he looked to determine the model of electric guitars. So there is a difference between it being a multi-classification model and ours being binary. His code is available on GitHub. Similar to ours, the images are converted to a Tensor with 3 color channels. He used an input shape of 224x224 and used transfer learning with a pretrained model. Dropout layers were used, early stopping, an activation function of “softmax”, and a loss function of “categorical cross-entropy”. The model was trained for 100 epochs but early stopping would stop training much before that. The end result was a mean accuracy score of 70-75% (Sengupta).

The second project “Guitar Segmentation in RGB Images Using Convolutional Neural Networks” was done by a group of college students from various european countries. Their CNN was to detect guitar and non-guitar images. They started with 2,200 RGB images and used preprocessing algorithms to create a dataset of 10,400 to divide 80/20 into training and validation. The chosen loss function was “categorical cross-entropy”. 20 epochs were run with a learning rate of 0.005 and batch size of 4. Transfer learning using various pretrained models was also used. In the end they ended with an mIoU (Mean Intersection Over Union) metric score of  95% using HRNet and 93% using Deeplab3+ (Tono).

The third project “Musical Instrument Recognition in User-generated Videos using a Multimodal Convolutional Neural Network Architecture” was done by a group of college students from Barcelona, Spain. Instead of just looking at types of guitars this project looks at recognizing musical instruments using video images and audio (we are only interested in the image part). They used a Stochastic Gradient Descent optimizer with an initial learning rate of 0.001, and  a momentum of 0.9. The learning rate was halved every 5 epochs. Early stopping of 5 epochs was used with a batch size of 64. Pre-training  used the Inception v3 model and model weights trained from ImageNet. The result being  71% for the performance of their classifier (Slizovskaia).

After looking at all three projects it seemed that the authors were satisfied with their results based on the specifics of the model used. Common in the projects was transfer learning using pre-trained models to beef up the image set and reduce the time to run the models, which is something we plan to do. While a lot of the details and complexities of these projects were beyond what we are doing, we found them very informative to review and learned from each. We feel it helped in the creation of our model.

# Basic CNN

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                    input_shape=(150, 150, 3)))
model.add(layers.Conv2D(32, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(256, (3, 3), activation='relu'))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
model.summary()
```

```python
from tensorflow.keras import optimizers

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

# Apply data augmentation
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,)

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')

validation_generator = val_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='binary')
```

```python
from keras.callbacks import EarlyStopping

es = EarlyStopping(monitor='val_loss', mode='min', 
          patience=10, 
          verbose=1,
          restore_best_weights=True)
```

```python
from tensorflow.keras import optimizers

model.compile(loss='binary_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
from keras.callbacks import EarlyStopping

es = EarlyStopping(monitor='val_loss', mode='min', 
          patience=10, 
          verbose=1,
          restore_best_weights=True)
```

```python
history = model.fit(
      train_generator,
      steps_per_epoch=70,
      epochs=100,
      validation_data=validation_generator,
      validation_steps=30,
      verbose=1,
      callbacks=[es])
```
